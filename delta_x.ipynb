```python
import numpy as np
import matplotlib.pyplot as pit 
```

Definizione della funziona $f(x)$

$f(x) = x^3 + x^2 - 5x + 2$

```python
def f(x):
    return(x**3 + x**2 - 5*x + 2)
```

```python
x = 0
f(x)
```

Genero il vettore degli argomenti $x$

```python
x = np.linspace(-10 , 10 , 201)
#print(x)
```

Genero il vettore delle immagini $y$

```python
# Grafico della funzione f(x) con limiti su y
import matplotlib.pyplot as plt
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Grafico di f(x) = x^3 + x^2 - 5x + 2')
plt.ylim(-10, 10)
plt.grid(True)
plt.show()
```

```python
def m(x):
    dx=1e-5
    return (f(x + dx) - f(x - dx)) / (2 * dx)
```

```python
# Calcolo la derivata numerica m(x) su tutto il vettore x
y_deriv = m(x)

plt.plot(x, f(x), color='hotpink', label='f(x)')
for dx in [0.1, 0.5, 1]:
    plt.plot(x, (f(x + dx) - f(x - dx)) / (2 * dx), label=f"m(x) con dx={dx}")
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x) e m(x) per x in [-10, 10]')
plt.ylim(-10, 10)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calcolo dei coefficienti del polinomio che approssima m(x)
coeff_m = np.polyfit(x, y_deriv, 2)
print('Coefficienti del polinomio approssimante m(x):', coeff_m)
```
